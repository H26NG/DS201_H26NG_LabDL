```python
!pip install underthesea
import re
import unicodedata

import pandas as pd

try:
    from underthesea import word_tokenize
    USE_UNDERTHESEA = True
except ImportError:
    print("underthesea chưa cài, sẽ fallback sang .split()")
    USE_UNDERTHESEA = False
```

```python
def normalize_unicode(text: str) -> str:
    """Chuẩn hóa Unicode về dạng NFC (chuẩn cho tiếng Việt)."""
    return unicodedata.normalize("NFC", str(text))


# pattern emoji & ký tự đặc biệt phổ biến
EMOJI_PATTERN = re.compile(
    "["
    "\U0001F600-\U0001F64F"  # emoticons
    "\U0001F300-\U0001F5FF"  # symbols & pictographs
    "\U0001F680-\U0001F6FF"  # transport & map symbols
    "\U0001F1E0-\U0001F1FF"  # flags
    "]+",
    flags=re.UNICODE,
)

URL_PATTERN = re.compile(r"http\S+|www\.\S+")


def clean_basic(text: str) -> str:
    """
    - Chuẩn hóa unicode
    - Lowercase
    - Bỏ URL, emoji
    - Bỏ ký tự thừa, chỉ giữ chữ, số, khoảng trắng, gạch dưới (để giữ token kiểu Giảng_viên)
    - Chuẩn hóa khoảng trắng
    """
    text = normalize_unicode(text)
    text = text.lower()

    # bỏ url
    text = URL_PATTERN.sub(" ", text)

    # bỏ emoji
    text = EMOJI_PATTERN.sub(" ", text)

    # bỏ ký tự thừa (giữ: chữ cái, số, khoảng trắng, gạch dưới)
    text = re.sub(r"[^0-9a-zA-ZÀ-Ỵà-ỹ_\s]", " ", text)

    # chuẩn hóa khoảng trắng
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
def tokenize_vi(text: str) -> str:
    """
    Tách từ tiếng Việt:
    - Nếu có underthesea: dùng word_tokenize -> token có gạch dưới: 'giảng_viên'
    - Nếu không: fallback sang split() thường
    """
    if USE_UNDERTHESEA:
        tokens = word_tokenize(text)       # trả về list
        # Ghép lại thành chuỗi, mỗi token cách nhau 1 space
        return " ".join(tokens)
    else:
        return " ".join(str(text).split())
```

```python
def preprocess_text(text: str) -> str:
    """
    Pipeline đầy đủ cho 1 câu:
    1. Clean cơ bản (unicode, lowercase, emoji, ký tự thừa, khoảng trắng)
    2. Tokenization tiếng Việt
    """
    text = clean_basic(text)
    text = tokenize_vi(text)
    return text
```

```python
for df in [df_train, df_dev, df_test]:
    df["sentence_clean"] = df["sentence"].astype(str).apply(preprocess_text)

df_train[["sentence", "sentence_clean"]].head()
```

### Bài 1: Xây dựng mạng LSTM gồm 5 lớp với hidden size là 256 cho bài toán phân loại văn bản. Huấn luyện mô hình này trên bộ dữ liệu UIT-VSFC (Vietnamese Student Feedback Corpus) sử dụng Adam làm phương thức tối ưu tham số và đánh giá độ hiệu quả của mô hình sử dụng độ đo F1.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.metrics import f1_score, classification_report

MAX_LEN = 50
MAX_VOCAB_SIZE = 20000
EMBED_DIM = 300
HIDDEN_SIZE = 256
NUM_LAYERS = 5
BATCH_SIZE = 32
EPOCHS = 10

# Topic labels 0..3
y_train = df_train["topic"].astype(int).values
y_dev   = df_dev["topic"].astype(int).values
y_test  = df_test["topic"].astype(int).values

NUM_CLASSES = len(np.unique(y_train))
LABELS = [str(i) for i in range(NUM_CLASSES)]

tokenizer = Tokenizer(num_words=MAX_VOCAB_SIZE, oov_token="<UNK>")
tokenizer.fit_on_texts(df_train["sentence_clean"].astype(str).tolist())

X_train = pad_sequences(tokenizer.texts_to_sequences(df_train["sentence_clean"].astype(str).tolist()),
                        maxlen=MAX_LEN, padding="post", truncating="post")
X_dev   = pad_sequences(tokenizer.texts_to_sequences(df_dev["sentence_clean"].astype(str).tolist()),
                        maxlen=MAX_LEN, padding="post", truncating="post")
X_test  = pad_sequences(tokenizer.texts_to_sequences(df_test["sentence_clean"].astype(str).tolist()),
                        maxlen=MAX_LEN, padding="post", truncating="post")

vocab_size = min(MAX_VOCAB_SIZE, len(tokenizer.word_index) + 1)
print("vocab_size:", vocab_size, "| X_train:", X_train.shape, "| classes:", NUM_CLASSES)
```

```python
class F1PerEpoch(tf.keras.callbacks.Callback):
    def __init__(self, X_val, y_val, average="macro", batch_size=32, name="val_f1"):
        super().__init__()
        self.X_val = X_val
        self.y_val = y_val
        self.average = average
        self.batch_size = batch_size
        self.name = name

    def on_epoch_end(self, epoch, logs=None):
        y_prob = self.model.predict(self.X_val, batch_size=self.batch_size, verbose=0)
        y_pred = np.argmax(y_prob, axis=1)
        f1 = f1_score(self.y_val, y_pred, average=self.average)
        print(f" — {self.name}: {f1:.2f}")
```

```python
def build_bilstm_5layers(vocab_size, num_classes):
    inputs = tf.keras.Input(shape=(MAX_LEN,), dtype=tf.int32)
    x = tf.keras.layers.Embedding(vocab_size, EMBED_DIM, mask_zero=True)(inputs)

    for _ in range(NUM_LAYERS - 1):
        x = tf.keras.layers.Bidirectional(
            tf.keras.layers.LSTM(HIDDEN_SIZE, return_sequences=True)
        )(x)

    x = tf.keras.layers.Bidirectional(
        tf.keras.layers.LSTM(HIDDEN_SIZE, return_sequences=False)
    )(x)

    x = tf.keras.layers.Dropout(0.3)(x)
    outputs = tf.keras.layers.Dense(num_classes, activation="softmax")(x)

    model = tf.keras.Model(inputs, outputs)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])
    return model

lstm_model = build_bilstm_5layers(vocab_size, NUM_CLASSES)
lstm_model.summary()

```

```python
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True),
    F1PerEpoch(X_dev, y_dev, average="macro", batch_size=BATCH_SIZE, name="val_f1_macro"),
    F1PerEpoch(X_dev, y_dev, average="weighted", batch_size=BATCH_SIZE, name="val_f1_weighted"),
]

history = lstm_model.fit(
    X_train, y_train,
    validation_data=(X_dev, y_dev),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks,
    verbose=1
)
```

```python
y_pred = np.argmax(lstm_model.predict(X_test, batch_size=BATCH_SIZE), axis=1)

print("\n===== FINAL TEST RESULTS =====")
print("Test F1 (macro)   :", f1_score(y_test, y_pred, average="macro"))
print("Test F1 (micro)   :", f1_score(y_test, y_pred, average="micro"))
print("Test F1 (weighted):", f1_score(y_test, y_pred, average="weighted"))
print("\nClassification report:")
print(classification_report(y_test, y_pred, target_names=LABELS, digits=2))

# Save
lstm_model.save("cau1_bilstm_5layers_256.keras")
print("Saved: cau1_bilstm_5layers_256.keras")
```

Kết quả theo F1 cho thấy mô hình làm tốt về tổng thể nhưng hiệu năng giữa các lớp không đồng đều do mất cân bằng nhãn. Trên tập test, F1 micro = 0.8594 và F1 weighted = 0.8553 khá cao (gần với accuracy ~0.86), điều này chủ yếu phản ánh mô hình dự đoán tốt các lớp có nhiều mẫu. Ngược lại, F1 macro = 0.7387 thấp hơn rõ rệt vì macro-F1 lấy trung bình đều trên các lớp, nên bị kéo xuống bởi các lớp ít mẫu và khó phân biệt.

Xét theo từng lớp: class 0 đạt F1 = 0.92 (precision 0.91, recall 0.94) với support rất lớn (2290), chứng tỏ mô hình học rất ổn lớp đa số. class 2 đạt F1 = 0.89 (precision 0.86, recall 0.94) dù chỉ có 145 mẫu, cho thấy lớp này có đặc trưng từ vựng/biểu đạt khá rõ nên mô hình bắt được tốt. class 1 có F1 = 0.68 (precision 0.76 nhưng recall 0.62), nghĩa là mô hình dự đoán class 1 khá “chặt” (ít đoán nhầm) nhưng lại bỏ sót nhiều mẫu true của lớp này (nhiều câu class 1 bị dự đoán sang lớp khác). class 3 là yếu nhất với F1 = 0.45 (precision 0.46, recall 0.45) và support 159, cho thấy mô hình chưa học tốt ranh giới của lớp này và hiệu năng bị ảnh hưởng bởi số mẫu ít; đây cũng là nguyên nhân chính khiến macro-F1 không cao. Tổng hợp lại, khoảng cách giữa weighted-F1 (0.8553) và macro-F1 (0.7387) là dấu hiệu điển hình của dữ liệu mất cân bằng và/hoặc độ khó phân biệt ở các lớp thiểu số.

### Bài 2: Xây dựng mạng GRU gồm 5 lớp với hidden size là 256 cho bài toán phân loại văn bản. Huấn luyện mô hình này trên bộ dữ liệu UIT-VSFC (Vietnamese Student Feedback Corpus) sử dụng Adam làm phương thức tối ưu tham số và đánh giá độ hiệu quả của mô hình sử dụng độ đo F1.

```python
def build_bigru_5layers(vocab_size, num_classes,
                        embed_dim=EMBED_DIM,
                        hidden_size=HIDDEN_SIZE,
                        num_layers=NUM_LAYERS,
                        max_len=MAX_LEN,
                        dropout=0.3):
    inputs = tf.keras.Input(shape=(max_len,), dtype=tf.int32)
    x = tf.keras.layers.Embedding(vocab_size, embed_dim, mask_zero=True)(inputs)

    # (num_layers - 1) lớp đầu: return_sequences=True
    for _ in range(num_layers - 1):
        x = tf.keras.layers.Bidirectional(
            tf.keras.layers.GRU(hidden_size, return_sequences=True)
        )(x)

    # lớp cuối: return_sequences=False
    x = tf.keras.layers.Bidirectional(
        tf.keras.layers.GRU(hidden_size, return_sequences=False)
    )(x)

    x = tf.keras.layers.Dropout(dropout)(x)
    outputs = tf.keras.layers.Dense(num_classes, activation="softmax")(x)

    model = tf.keras.Model(inputs, outputs)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model
```

```python
gru_model = build_bigru_5layers(vocab_size, NUM_CLASSES)
gru_model.summary()
```

```python
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True),
    F1PerEpoch(X_dev, y_dev, average="macro", batch_size=BATCH_SIZE, name="val_f1_macro"),
    F1PerEpoch(X_dev, y_dev, average="weighted", batch_size=BATCH_SIZE, name="val_f1_weighted"),
]

history = gru_model.fit(
    X_train, y_train,
    validation_data=(X_dev, y_dev),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks,
    verbose=1
)
```

```python
y_pred = np.argmax(gru_model.predict(X_test, batch_size=BATCH_SIZE), axis=1)

print("\n===== FINAL TEST RESULTS (CÂU 2 - BiGRU) =====")
print("Test F1 (macro)   :", f1_score(y_test, y_pred, average="macro"))
print("Test F1 (micro)   :", f1_score(y_test, y_pred, average="micro"))
print("Test F1 (weighted):", f1_score(y_test, y_pred, average="weighted"))
print("\nClassification report:")
print(classification_report(y_test, y_pred, target_names=LABELS, digits=2))

gru_model.save("cau2_bigru_5layers_256.keras")
print("Saved: cau2_bigru_5layers_256.keras")
```

Kết quả GRU cho thấy mô hình tổng thể nhỉnh hơn LSTM một chút ở thước đo công bằng giữa các lớp. Trên tập test, F1 macro = 0.7419, F1 micro = 0.8636, F1 weighted = 0.8625; so với LSTM (macro ~0.7387) thì macro-F1 tăng nhẹ, còn micro/weighted cũng cao và tương ứng với accuracy ~0.86. Như vậy mô hình vẫn dự đoán tốt lớp đa số, đồng thời cải thiện phần nào ở các lớp thiểu số.

Theo từng lớp, class 0 đạt F1 = 0.92 (precision 0.95, recall 0.90) rất tốt nhưng recall thấp hơn một chút, tức mô hình “khắt khe hơn” khi dự đoán class 0 (ít đoán nhầm vào class 0). class 1 đạt F1 = 0.73 với recall rất cao 0.86 nhưng precision 0.64, nghĩa là mô hình bắt được nhiều mẫu class 1 (ít bỏ sót), đổi lại có xu hướng dự đoán nhầm sang class 1 nhiều hơn. class 2 vẫn tốt với F1 = 0.90 (precision 0.94, recall 0.87), ổn định dù số mẫu ít. class 3 vẫn là lớp khó nhất với F1 = 0.41 (precision 0.65, recall 0.30): mô hình dự đoán class 3 khá “chặt” (precision cao) nhưng bỏ sót nhiều mẫu class 3 (recall thấp), nên F1 còn thấp và đây là yếu tố kéo macro-F1 xuống.

Về quá trình train, val_f1_macro đạt khoảng 0.73 từ epoch 2–4 rồi giảm ở epoch 5, đồng thời val_loss tăng về sau, cho thấy mô hình đạt điểm tốt sớm và có dấu hiệu overfit nhẹ nếu train quá lâu; chọn checkpoint/early stopping quanh epoch 2–4 là hợp lý. Tổng kết lại, BiGRU là lựa chọn tốt hơn BiLSTM trong thử nghiệm này vì macro-F1 cao hơn và class 1 được cải thiện rõ, nhưng class 3 vẫn là điểm nghẽn chính do mất cân bằng nhãn và khó phân biệt, cần các kỹ thuật như class_weight/oversampling hoặc tuning regularization để tăng recall cho lớp này.

```python
!pip -q install seqeval

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from seqeval.metrics import f1_score as ner_f1
from seqeval.metrics import classification_report as ner_report

#CONFIG
MAX_LEN = 80
EMBED_DIM = 300
HIDDEN_SIZE = 256
NUM_LAYERS = 5
BATCH_SIZE = 32
EPOCHS = 20
DROPOUT = 0.3
```

```python
TRAIN_PATH = "/content/train_word.conll"
DEV_PATH   = "/content/dev_word.conll"
TEST_PATH  = "/content/test_word.conll"
```

```python
def read_conll(path):
    sentences, tags = [], []
    cur_tokens, cur_tags = [], []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                if cur_tokens:
                    sentences.append(cur_tokens)
                    tags.append(cur_tags)
                    cur_tokens, cur_tags = [], []
                continue
            parts = line.split()
            cur_tokens.append(parts[0])
            cur_tags.append(parts[-1])
    if cur_tokens:
        sentences.append(cur_tokens)
        tags.append(cur_tags)
    return sentences, tags

train_sents, train_tags = read_conll(TRAIN_PATH)
dev_sents, dev_tags     = read_conll(DEV_PATH)
test_sents, test_tags   = read_conll(TEST_PATH)

print("Train/Dev/Test:", len(train_sents), len(dev_sents), len(test_sents))
```

```python
from collections import Counter

def build_token_vocab(sentences, min_freq=1):
    cnt = Counter()
    for sent in sentences:
        cnt.update(sent)
    token2id = {"<PAD>": 0, "<UNK>": 1}
    for tok, c in cnt.most_common():
        if c >= min_freq and tok not in token2id:
            token2id[tok] = len(token2id)
    id2token = {i:t for t,i in token2id.items()}
    return token2id, id2token

def build_tag_vocab(tag_seqs):
    tags = sorted({t for seq in tag_seqs for t in seq})
    tag2id = {"<PAD>": 0}
    for t in tags:
        if t not in tag2id:
            tag2id[t] = len(tag2id)
    id2tag = {i:t for t,i in tag2id.items()}
    return tag2id, id2tag

token2id, id2token = build_token_vocab(train_sents, min_freq=1)
tag2id, id2tag     = build_tag_vocab(train_tags)

print("Vocab size:", len(token2id))
print("Num tags  :", len(tag2id))
```

```python
def encode_sentences(sentences, token2id):
    unk = token2id["<UNK>"]
    return [[token2id.get(tok, unk) for tok in sent] for sent in sentences]

def encode_tags(tag_seqs, tag2id):
    return [[tag2id[t] for t in seq] for seq in tag_seqs]

X_train = pad_sequences(encode_sentences(train_sents, token2id), maxlen=MAX_LEN, padding="post", truncating="post", value=0)
X_dev   = pad_sequences(encode_sentences(dev_sents, token2id),   maxlen=MAX_LEN, padding="post", truncating="post", value=0)
X_test  = pad_sequences(encode_sentences(test_sents, token2id),  maxlen=MAX_LEN, padding="post", truncating="post", value=0)

y_train = pad_sequences(encode_tags(train_tags, tag2id), maxlen=MAX_LEN, padding="post", truncating="post", value=0)
y_dev   = pad_sequences(encode_tags(dev_tags, tag2id),   maxlen=MAX_LEN, padding="post", truncating="post", value=0)
y_test  = pad_sequences(encode_tags(test_tags, tag2id),  maxlen=MAX_LEN, padding="post", truncating="post", value=0)

# sparse CE cần shape (B,T,1)
y_train_ce = np.expand_dims(y_train, -1)
y_dev_ce   = np.expand_dims(y_dev, -1)
```

```python
def build_ner_encoder(vocab_size, num_tags):
    inputs = tf.keras.Input(shape=(MAX_LEN,), dtype=tf.int32)
    x = tf.keras.layers.Embedding(vocab_size, EMBED_DIM, mask_zero=True)(inputs)

    for _ in range(NUM_LAYERS):
        x = tf.keras.layers.Bidirectional(
            tf.keras.layers.LSTM(
                HIDDEN_SIZE,
                return_sequences=True,
                dropout=DROPOUT
            )
        )(x)

    outputs = tf.keras.layers.Dense(num_tags, activation="softmax")(x)
    model = tf.keras.Model(inputs, outputs)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-3),
        loss="sparse_categorical_crossentropy"
    )
    return model

model = build_ner_encoder(vocab_size=len(token2id), num_tags=len(tag2id))
model.summary()
```

```python
def eval_seqeval(model, X, y_true_ids, id2tag, pad_id=0):
    y_prob = model.predict(X, batch_size=BATCH_SIZE, verbose=0)
    y_pred_ids = np.argmax(y_prob, axis=-1)

    true_tags, pred_tags = [], []
    for t_seq, p_seq in zip(y_true_ids, y_pred_ids):
        t_out, p_out = [], []
        for t, p in zip(t_seq, p_seq):
            if int(t) == pad_id:
                break
            t_out.append(id2tag[int(t)])
            p_out.append(id2tag[int(p)])
        true_tags.append(t_out)
        pred_tags.append(p_out)

    f1 = ner_f1(true_tags, pred_tags)
    rep = ner_report(true_tags, pred_tags, digits=4)
    return f1, rep
```

```python
class NERF1Callback(tf.keras.callbacks.Callback):
    def __init__(self, X_val, y_val, id2tag, name="val_f1"):
        super().__init__()
        self.X_val = X_val
        self.y_val = y_val
        self.id2tag = id2tag
        self.name = name

    def on_epoch_end(self, epoch, logs=None):
        f1, _ = eval_seqeval(self.model, self.X_val, self.y_val, self.id2tag)
        print(f" — {self.name}: {f1:.4f}")

callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=2, restore_best_weights=True),
    NERF1Callback(X_dev, y_dev, id2tag, name="val_f1_entity"),
]
```

```python
history = model.fit(
    X_train, y_train_ce,
    validation_data=(X_dev, y_dev_ce),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks,
    verbose=1
)
```

```python
def build_ner_encoder(vocab_size, num_tags):
    inputs = tf.keras.Input(shape=(MAX_LEN,), dtype=tf.int32)
    x = tf.keras.layers.Embedding(vocab_size, EMBED_DIM, mask_zero=True)(inputs)

    for _ in range(NUM_LAYERS):
        x = tf.keras.layers.Bidirectional(
            tf.keras.layers.LSTM(
                HIDDEN_SIZE,
                return_sequences=True,
                dropout=DROPOUT
            )
        )(x)

    outputs = tf.keras.layers.Dense(num_tags, activation="softmax")(x)
    model = tf.keras.Model(inputs, outputs)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-3),
        loss="sparse_categorical_crossentropy"
    )
    return model

model = build_ner_encoder(vocab_size=len(token2id), num_tags=len(tag2id))
model.summary()
```

```python
def eval_seqeval(model, X, y_true_ids, id2tag, pad_id=0):
    y_prob = model.predict(X, batch_size=BATCH_SIZE, verbose=0)
    y_pred_ids = np.argmax(y_prob, axis=-1)

    true_tags, pred_tags = [], []
    for t_seq, p_seq in zip(y_true_ids, y_pred_ids):
        t_out, p_out = [], []
        for t, p in zip(t_seq, p_seq):
            if int(t) == pad_id:
                break
            t_out.append(id2tag[int(t)])
            p_out.append(id2tag[int(p)])
        true_tags.append(t_out)
        pred_tags.append(p_out)

    f1 = ner_f1(true_tags, pred_tags)
    rep = ner_report(true_tags, pred_tags, digits=4)
    return f1, rep
```

```python
class NERF1Callback(tf.keras.callbacks.Callback):
    def __init__(self, X_val, y_val, id2tag, name="val_f1"):
        super().__init__()
        self.X_val = X_val
        self.y_val = y_val
        self.id2tag = id2tag
        self.name = name

    def on_epoch_end(self, epoch, logs=None):
        f1, _ = eval_seqeval(self.model, self.X_val, self.y_val, self.id2tag)
        print(f" — {self.name}: {f1:.4f}")

callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=2, restore_best_weights=True),
    NERF1Callback(X_dev, y_dev, id2tag, name="val_f1_entity"),
]
```

```python
history = model.fit(
    X_train, y_train_ce,
    validation_data=(X_dev, y_dev_ce),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks,
    verbose=1
)
```

```python
test_f1, rep = eval_seqeval(model, X_test, y_test, id2tag)
print("\n FINAL TEST (NER)")
print("Entity-level F1:", test_f1)
print(rep)

model.save("phoner_encoder_syllable_bilstm5_256.keras")
```

Theo macro-F1 (trung bình đều theo từng loại thực thể), mô hình word-level cho kết quả tốt và “công bằng” hơn so với syllable-level: macro F1 = 0.7550 (word) so với 0.7112 (syllable). Vì macro-F1 không bị chi phối bởi các thực thể phổ biến, khoảng cách này cho thấy mô hình word-level xử lý các nhãn khó/hiếm tốt hơn và ổn định hơn. Trong cả hai mô hình, các nhãn phổ biến và có đặc trưng rõ như DATE, LOCATION, PATIENT_ID vẫn đạt F1 cao, nhưng các nhãn khó như JOB và NAME có F1 thấp đã kéo macro-F1 xuống (đặc biệt ở syllable-level). Nguyên nhân hợp lý là tokenization theo word giữ được đơn vị nghĩa rõ ràng và chuỗi ngắn hơn, giúp BiLSTM học ranh giới thực thể tốt hơn; còn syllable tạo chuỗi dài và dễ nhiễu, khiến việc nhận diện chính xác các thực thể hiếm trở nên khó hơn.
